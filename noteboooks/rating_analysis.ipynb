# 1. IMPORT LIBRARIES

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from wordcloud import WordCloud

import warnings
warnings.filterwarnings("ignore")
```

# 2. Load dataset

```python
df = pd.read_csv("../data/raw/zomato.csv")
```

# 3. Basic Data Overview

```python
print("\nFirst 5 Rows")
print(df.head())

print("\nDataset Shape")
print(df.shape)

print("\nDataset Info")
print(df.info())

print("\nMissing values")
print(df.isnull().sum())

```

# 4. Remove Duplicates

```python
print("\nDuplicate Rows:", df.duplicated().sum())

df.drop_duplicates(inplace=True)

print("Duplicates Removed Successfully")

```

# 5. Clean rate column

```python

# Remove invalid values
df = df[df['rate'] != 'NEW']
df = df[df['rate'] != '-']

# Convert everything to string first
df['rate'] = df['rate'].astype(str)

# Remove '/5' safely
df['rate'] = df['rate'].str.replace('/5', '', regex=False)

# Convert to float
df['rate'] = pd.to_numeric(df['rate'], errors='coerce')

# Remove null values created during conversion
df.dropna(subset=['rate'], inplace=True)

print("\nRate Column Cleaned Successfully")
```

# 7. Clean cost column 

```python

# Convert column to string
df['approx_cost(for two people)'] = \
df['approx_cost(for two people)'].astype(str)

# Remove commas safely
df['approx_cost(for two people)'] = \
df['approx_cost(for two people)'].str.replace(',', '', regex=False)

# Convert to numeric
df['approx_cost(for two people)'] = pd.to_numeric(
    df['approx_cost(for two people)'],
    errors='coerce'
)

# Remove null values
df.dropna(
    subset=['approx_cost(for two people)'],
    inplace=True
)

print("\nCost Column Cleaned Successfully")
```

# 8. Handle missing values

```python
print("\nMissing Values Before Cleaning")
print(df.isnull().sum())

df.dropna(inplace=True)

print("\nMissing Values After Cleaning")
print(df.isnull().sum())
```

# 9. Final dataset shape

```python
print("\nFinal Dataset Shape")
print(df.shape)
```

# 10. Save cleaned dataset 

```python
output_path = "../data/processed/zomato_cleaned.csv"

df.to_csv(output_path, index=False)

print(f"\nCleaned dataset saved successfully at:\n{output_path}")

```
