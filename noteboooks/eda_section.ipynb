# EDA section

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from wordcloud import WordCloud

df = pd.read_csv("../data/processed/zomato_cleaned.csv")
```

```python
sns.set_style("darkgrid")
```

# 11. Rating distribution

```python
plt.figure(figsize=(10,6))

sns.histplot(
    df['rate'],
    bins=20,
    kde=True
)

plt.title("Restaurant Rating Distribution")

plt.xlabel("Ratings")

plt.ylabel("Count")

plt.show()
```

# 12. Online order analysis

```python
plt.figure(figsize=(7,5))

sns.countplot(
    x='online_order',
    data=df
)

plt.title("Online Order Availability")

plt.xlabel("Online Order")

plt.ylabel("Count")

plt.show()
```

# 13. table booking analysis

```python
plt.figure(figsize=(7,5))

sns.countplot(
    x='online_order',
    data=df
)

plt.title("Online Order Availability")

plt.xlabel("Online Order")

plt.ylabel("Count")

plt.show()
```

# 14. Top 10 locations 

```python
top_locations = df['location'].value_counts().head(10)

plt.figure(figsize=(12,6))

sns.barplot(
    x=top_locations.index,
    y=top_locations.values
)

plt.title("Top 10 Restaurant Locations")

plt.xlabel("Location")

plt.ylabel("Number of Restaurants")

plt.xticks(rotation=45)

plt.show()
```

# 15. top 10 cuisines

```python
top_cuisines = df['cuisines'].value_counts().head(10)

plt.figure(figsize=(12,6))

sns.barplot(
    x=top_cuisines.index,
    y=top_cuisines.values
)

plt.title("Top 10 Cuisines")

plt.xlabel("Cuisine")

plt.ylabel("Count")

plt.xticks(rotation=75)

plt.show()
```

# 16. Cost Distribution

```python
plt.figure(figsize=(10,6))

sns.histplot(
    df['approx_cost(for two people)'],
    bins=30,
    kde=True
)

plt.title("Cost Distribution")

plt.xlabel("Approx Cost for Two People")

plt.ylabel("Count")

plt.show()
```

# 17. votes vs ratings

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    x='votes',
    y='rate',
    data=df
)

plt.title("Votes vs Ratings")

plt.xlabel("Votes")

plt.ylabel("Ratings")

plt.show()
```

# 18. top restaurant types 

```python
top_rest_types = df['rest_type'].value_counts().head(10)

plt.figure(figsize=(12,6))

sns.barplot(
    x=top_rest_types.index,
    y=top_rest_types.values
)

plt.title("Top Restaurant Types")

plt.xlabel("Restaurant Type")

plt.ylabel("Count")

plt.xticks(rotation=60)

plt.show()
```

# 19. most voted restaurants 

```python
top_voted = df.sort_values(
    by='votes',
    ascending=False
).head(10)

plt.figure(figsize=(12,6))

sns.barplot(
    x=top_voted['name'],
    y=top_voted['votes']
)

plt.title("Top 10 Most Voted Restaurants")

plt.xlabel("Restaurant")

plt.ylabel("Votes")

plt.xticks(rotation=75)

plt.show()
```

 # 20. HEATMAP

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    df[['rate', 'votes',
        'approx_cost(for two people)']].corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")

plt.show()
```

# 21. word cloud

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Use only first 5000 reviews for faster processing
text = " ".join(
    df['reviews_list']
    .dropna()
    .astype(str)
    .head(5000)
)

# Generate wordcloud
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white',
    max_words=200
).generate(text)

# Plot
plt.figure(figsize=(12,6))

plt.imshow(wordcloud, interpolation='bilinear')

plt.axis('off')

plt.title("Most Frequent Review Words")

plt.show()
```
